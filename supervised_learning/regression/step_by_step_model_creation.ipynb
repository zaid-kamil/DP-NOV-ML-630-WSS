```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
from statsmodels.regression.linear_model import OLS
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/regression/50_Startups.csv')
```

```python
X = df.drop('Profit', axis=1)
y = df['Profit']
```

```python
state_enc = OneHotEncoder(drop='first')
dummy_state = state_enc.fit_transform(X[['State']]).toarray()

X = X.drop('State', axis=1)
const = np.ones((X.shape[0], 1))
X = pd.concat([pd.DataFrame(const), X, pd.DataFrame(dummy_state)], axis=1)
```

```python
model = OLS(y, X)
results = model.fit()
results.summary()
```

```python
Xo = X.drop(1, axis=1)
model = OLS(y, Xo)
results = model.fit()
results.summary()
```

```python
Xo = Xo.drop(0, axis=1)
model = OLS(y, Xo)
results = model.fit()
results.summary()
```
