```python
def milage_bin(val):
    if val < 20:
        return "Low Mileage"
    elif val < 35:
        return "Medium Mileage"
    else:
        return "High Mileage"
```

```python
df['mileage-group'] = df['city-mpg'].apply(milage_bin)
df.head(3)
```
